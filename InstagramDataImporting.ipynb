### Liked Comments

```python
with open('liked_comments.json', 'r') as file:
    Liked_Comments = json.load(file)

# 'data' now contains the JSON data as a Python dictionary
print(Liked_Comments)
```

```python
Result = []
TitleHolder = []
HrefHolder = []
ValueHolder = []
TimeStampHolder = []

for item in Liked_Comments['likes_comment_likes']:
    title = item['title']
    for string_data in item['string_list_data']:
        href = string_data['href']
        value = string_data['value']
        timestamp = string_data['timestamp']
        entry = {"Title":title, "Href":href, "Value": value, "Timestamp":timestamp}
        Result.append(entry)
        TitleHolder.append(title)
        HrefHolder.append(href)
        ValueHolder.append(value)
        TimeStampHolder.append(timestamp)
    
        print(f"Title: {title}, Href: {href}, Value: {value}, Timestamp: {timestamp}")
```

```python
df = pd.DataFrame(Result)
```

### Actual Pre-Processing

#### SavedPosts

```python
SavedPosts
```

```python
SavedPosts['Actual Time'] = SavedPosts['Timestamp'].apply(lambda x: datetime.datetime.fromtimestamp(x))
SavedPosts['Actual Time in String'] = SavedPosts['Actual Time'].apply(lambda x: x.strftime("%Y-%m-%d %H:%M:%S"))
```

```python
SavedPosts['Month'] = SavedPosts['Actual Time in String'].apply(lambda x: x[5:7])
SavedPosts['Hour'] = SavedPosts['Actual Time in String'].apply(lambda x: x[11:13])
```

```python
SavedPosts['EST Military Hour'] = SavedPosts['Hour'].apply(lambda x: int(x))

SavedPosts['EST Standard Hour'] = SavedPosts['EST Military Hour'].apply(lambda x: str(x) + ' AM' if x < 12 else str(x - 12) + ' PM')

SavedPosts['Numeric EST Standard Hour'] = SavedPosts['EST Standard Hour'].apply(lambda x: x[:1])

bins = [0,8,16,24]
labels = ['0-8','9-16','17-24']
SavedPosts['Category_Military_Hour'] = pd.cut(SavedPosts['EST Military Hour'], bins=bins, labels=labels)

SavedPosts
```
