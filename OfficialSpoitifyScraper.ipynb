```python
# Find the table element
table = driver.find_element(By.XPATH,'/html/body/div[3]/div[2]/div/div[2]/div/div/div[1]/table/div/div/table')
```

```python
# Read the table into a DataFrame
df = pd.read_html(table.get_attribute('outerHTML'))[0]
```
