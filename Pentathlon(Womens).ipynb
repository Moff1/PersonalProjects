```python
PentathlonWomensLink = "http://www.ustfccca.org/records-lists/outdoor-collegiate-all-time-bests?gender=2&sport=1&div=20527&top=25&event=30"
```

```python
page = requests.get(PentathlonWomensLink)
soup = BeautifulSoup(page.text,'html')
soup

table = soup.find("table")
```

```python
Header = soup.findAll('th')
Headers = []
for h in Header:
    Headers.append(h.text)
```

```python
row_soup = soup.findAll('tr')
```

```python
rows = []
for row in row_soup:
    row_data = []
    for cell in row.findAll('td'):
        row_data.append(cell.text)
    rows.append(row_data)
```

```python
RealRows = []
```

```python
for i in range(0,227):
    if len(rows[i]) > 1:
        RealRows.append(rows[i])
        
```

```python
RealRealRows = RealRows[::2]
RealRealRows
```

```python
len(RealRealRows)
```

```python
OfficialScores = []
```

```python
for i in range(0,len(RealRealRows)):
    OfficialScores.append(RealRealRows[i][4])
OfficialScores
```

```python
Scores = []
for i in range(0,len(OfficialScores)):
    Scores.append(OfficialScores[i][:5])
    
Scores
```

```python
OfficialNames = []
for i in range(0,len(RealRealRows)):
    OfficialNames.append(RealRealRows[i][1])
OfficialNames
```

```python
OfficialSchool = []
for i in range(0,len(RealRealRows)):
    OfficialSchool.append(RealRealRows[i][3])
OfficialSchool
```

```python
OfficialReGexNames = []
for i in range(0,len(RealRealRows)):
    if OfficialNames[i].startswith("\xa0\xa0"):
        OfficialReGexNames.append(OfficialNames[i][4:])
    else:
        OfficialReGexNames.append(OfficialNames[i])
        
OfficialReGexNames
```

```python
CleanedNames = []
for i in range(0,len(RealRealRows)):
    if OfficialReGexNames[i].startswith("—"):
        CleanedNames.append(OfficialReGexNames[i][2:])
    if OfficialReGexNames[i].startswith("("):
        CleanedNames.append(OfficialReGexNames[i][4:])        
CleanedNames
```

```python
CleanedNames2 = []
for i in range(0,len(RealRealRows)):
    if CleanedNames[i].startswith(" "):
        CleanedNames2.append(CleanedNames[i][1:])
    else:
        CleanedNames2.append(CleanedNames[i])        
CleanedNames2
```

```python
WomensPentDF = pd.DataFrame({'Names': CleanedNames2, 'Scores': Scores, 'Schools': OfficialSchool})
WomensPentDF.to_csv('WomensPent.csv', index=False)
```
