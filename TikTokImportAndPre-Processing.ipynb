### Pre-Processing

#### VBHIFLOT

```python
VBHIFLOT
```

```python
VBHIFLOT.rename(columns={'Date_x': 'Date', 'Link_x': 'Link','Day_x':'Day'}, inplace=True)
```

```python
VBHIFLOT = VBHIFLOT.drop(['Date_y','Link_y','Day_y'],axis = 1)
VBHIFLOT
```

```python
VBHIFLOT['Month'] = VBHIFLOT['Day'].apply(lambda x: x[5:7])
VBHIFLOT['Hour'] = VBHIFLOT['Time'].apply(lambda x: x[0:2])
VBHIFLOT['EST Military Hour'] = VBHIFLOT['Hour'].apply(lambda x: int(x))
VBHIFLOT['EST Standard Hour'] = VBHIFLOT['EST Military Hour'].apply(lambda x: str(x) + ' AM' if x < 12 else str(x - 12) + ' PM')
VBHIFLOT['Numeric EST Standard Hour'] = VBHIFLOT['EST Standard Hour'].apply(lambda x: x[:1])
```

```python
bins = [0,8,16,24]
labels = ['0-8','9-16','17-24']
VBHIFLOT['Category_Military_Hour'] = pd.cut(VBHIFLOT['EST Military Hour'], bins=bins, labels=labels)
```

```python
VBHIFLOT
```
